```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
```

```python
#load the dataset
census = pd.read_csv('DS/census.csv')
census.columns = ['ID','education','age','capital-gain', 'race', 
                  'capital-loss','hours-per-week','gender', 'class' ]
# Print the head of census
census.head()
```

```python
census['class'].value_counts()
```

```python
census.education = census.education.astype('category').cat.codes
census.race = census.race.astype('category').cat.codes
census.gender = census.gender.astype('category').cat.codes
```

```python
census['class'] = census['class'].astype('category').cat.codes
```

```python
census.dtypes
```

```python
census['capital-gain'] = pd.to_numeric(census['capital-gain'], errors='coerce')
```

```python
census.dtypes
```

```python
census
```

```python
census.isnull().sum()
```

```python
census['capital-gain'].fillna(census['capital-gain'].mean(), inplace=True)
```

```python
census.isnull().sum()
```

```python
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(census[census.columns[:-1]], census['class'])
```
