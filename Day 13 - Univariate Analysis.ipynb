# Box Plot

```python
airquality.Ozone.plot.box()
```

```python
plt.figure(figsize=(8, 6))

# Plotting the box plot for Ozone with customization
plt.boxplot(airquality['Ozone'].dropna(), 
            patch_artist=True,  # This will allow us to customize the box color
            boxprops=dict(facecolor='#3498db', color='black', linewidth=1.5),  # Blue box with black edges
            medianprops=dict(color='red', linewidth=2),  # Red median line
            whiskerprops=dict(color='blue', linewidth=1.5),  # Black whiskers
            capprops=dict(color='red', linewidth=1.5),  # Black caps
            flierprops=dict(marker='o', markerfacecolor='green', markersize=7, linestyle='none')  # Customize outliers
           )

# Add grid
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

# Customizing labels and title
plt.title('Box Plot of Ozone Levels', fontsize=16, fontweight='bold')
plt.ylabel('Ozone Levels', fontsize=12)
plt.xticks([1], ['Ozone'], fontsize=12)

# Show the plot
plt.show()
```

# Bar Chart

```python
df = pd.read_csv("Spotify_Youtube_Sample.csv")
df.head()
```

```python
df.describe(include='object')
```

```python
df.Album_type.value_counts().plot.bar()
```

```python
album_counts = df['Album_type'].value_counts()

# Set figure size
plt.figure(figsize=(7, 5))

# Plot bar chart
bars = plt.bar(album_counts.index, album_counts.values,color=['#3498db', '#2ecc71', '#e74c3c'])

# Customizing the plot
plt.title('Album Types Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Album Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)

# Adding gridlines
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Annotating bars with values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```

# Pie plot

```python
df.Album_type.value_counts().plot.pie()
```

```python
album_counts = df['Album_type'].value_counts()

# Set colors for the pie chart
colors = ['#3498db', '#2ecc71', '#e74c3c']

# Create a pie chart
plt.figure(figsize=(7, 7))

# Plot pie chart with customization
plt.pie(album_counts.values, 
        labels=album_counts.index, 
        autopct='%1.1f%%',     # Display percentages
        startangle=90,         # Start pie chart from 90 degrees
        colors=colors,         # Set custom colors
        wedgeprops={'edgecolor': 'black', 'linewidth': 1},  # Add border to wedges
        textprops={'fontsize': 11})  # Customize text size

# Title for the chart
plt.title('Album Types Distribution', fontsize=16, fontweight='bold')

# Show plot
plt.show()
```
