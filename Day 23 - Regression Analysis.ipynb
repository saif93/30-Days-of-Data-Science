# Video Example

```python
### Plotting the scatter diagram and the regression line.
import numpy as np
import matplotlib.pyplot as plt
y = np.array([64,87,50,71,44,56,42,60])
x = np.array([5,2,12,9,15,6,25,16])
plt.scatter(x,y)
```

### Our estimated regression line y = a + bx is:
### y = 76.66 - 1.5476 x

### Interpret the meaning of the values of a and b calculated
<ul><li>The value of a = 76.6605 gives the value of ŷ for x = 0; that is, it gives the monthly auto
insurance premium for a driver with no driving experience. </li>
<li>The value of b gives the change in ŷ due to a change of one unit in x. </li>
<li>Thus, b = −1.5476 indicates that, on average, for every extra year of driving experience, the monthly auto insurance premium decreases by $1.55. Note that when b is negative, y decreases as x increases.</li>

```python
### Plotting the scatter diagram and the regression line.

yp = [76.66 - 1.5476 * i for i in x]

# Plotting the scatter diagram
plt.scatter(x, y, label="Data Points")
plt.plot(x, yp, color='g', label="Regression Line")
plt.xlabel("Driving Experience")
plt.ylabel("Monthly Auto Insurance Premium")
plt.title("Scatter Plot with Regression Line")
plt.legend()
plt.show()
```

# Implementation of LR using scikit-learn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
marks = pd.read_csv('marks.csv')
```

```python
marks.plot.scatter('MID','FINAL')
```

```python
#separating dependent and independent variables
X = marks.MID.fillna(marks.MID.mean())
y = marks.FINAL.fillna(marks.FINAL.mean())
#Data Preparation
X = X.values.reshape(-1,1)
y= y.values.reshape(-1,1)
```

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# Model initialization
regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(x_train, y_train)
```

```python
# printing values
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
```

```python
# Predict
y_train_predict = regression_model.predict(x_train)
y_predicted = regression_model.predict(x_test)
```

```python
r2 = r2_score(y_train, y_train_predict)
print('R2 score: ', r2)
```

```python
#evaluation
mse_train = mean_squared_error(y_train, y_train_predict)
mse_test = mean_squared_error(y_test, y_predicted)
print('Mean squared error Training: ', mse_train)
print('Mean squared error Testing: ', mse_test)
print('Root mean squared error Training: ', mse_train**(1/2.0))
print('Root mean squared error Testing: ', mse_test**(1/2.0))
```

```python
# Plot outputs
import matplotlib.pyplot as plt
plt.scatter(x_test, y_test, color="black")
plt.plot(x_test, y_predicted, color="blue", linewidth=3)
```
